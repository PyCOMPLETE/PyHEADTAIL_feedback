```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

N=1000000

d_normal = np.random.randn(N)
d_uniform = 1./0.577263*(-1.+2.*np.random.rand(N))

plt.plot(d_normal,'b.')
plt.plot(d_uniform,'r.')

RMS_normal = np.sqrt(np.mean(d_normal*d_normal))
RMS_uniform = np.sqrt(np.mean(d_uniform*d_uniform))

print 'RMS_normal: ' + str(RMS_normal) + '    RMS_uniform: ' + str(RMS_uniform)

plt.show()


```
